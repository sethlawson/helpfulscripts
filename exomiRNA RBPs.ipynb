```python
#iterate through folder of excel files for postar clips of miRNAs
#in each given WB/sheet
    #collect all unique names in column 1, associate it with the name of the miRNA
import os, pprint

print('Using POSTAR PAR-CLIP data for top HBEC exomiRNAs. \nThis PAR-CLIP data is agnostic to the sense of the strand, therefore duplicate miR-199A-1 and 132 were removed.\n')

miRNA_dict = {}
for filename in os.listdir(r'C:\Users\lawsonsa\Desktop\C2C2 sequencing data\POSTAR exports of miRNA PAR-CLIP hits'):    
    df = pd.read_csv('C:\\Users\\lawsonsa\\Desktop\\C2C2 sequencing data\\POSTAR exports of miRNA PAR-CLIP hits\\' + filename)
    miRNA_dict[filename] = []
    for cell in df.iloc[:, 0]:
        if cell not in miRNA_dict[filename]:
            miRNA_dict[filename].append(cell)
            
pprint.pprint(miRNA_dict)
```

```python
#unique RBPs that came out with PAR-CLIP data
unique_RBPs = []
for key in miRNA_dict:
    for list_item in miRNA_dict[key]:
        if list_item not in unique_RBPs:
            unique_RBPs.append(list_item)
print('Unique RBPs found to bind with these exomiRNAs\n')            
pprint.pprint(unique_RBPs)

```

```python
#read mRNA array data WB/sheet
#for each key/miRNA in miRNA_dict
    #for each list item 
        #if key exists in this WB
            #collect all rows with that key in some object
        #else
            #print dictionary key that failed

#create a new excel sheet of miRNAs and corresponding RBP gene expression from the object above
mRNA_Array_df = pd.read_excel(r'C:\Users\lawsonsa\Desktop\C2C2 sequencing data\HBEC and 4017 cell mRNA data.xlsx', 0)    
listy = mRNA_Array_df.values.tolist()
print(listy)
for key in miRNA_dict:
    wb = Workbook()
    sheet = wb.create_sheet()
    dest_filename = 'C:\\Users\\lawsonsa\\Desktop\\C2C2 sequencing data\\' + key[0:-4] + ' binding RBPs and their expression data.xlsx'
    for item in miRNA_dict[key]:
        try:
            r = 2
            for i in listy:
                if i[0] == item:
                    for col, val in enumerate(i, start=1):
                        sheet.cell(row=r, column=col).value = val
                r += 1            
        except KeyError:
            print('No mRNA array data found for key ' + item)
    wb.save(filename = dest_filename)

    
```
